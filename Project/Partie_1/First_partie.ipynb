# Perceptron Simple

Dans cette section, on va développer un perceptron simple avec rétropropagation. 

<h3> Plan de travail </h3>

**Première partie :** On teste le perceptron sur des données synthétiques simples
- Données : `x = [[0,0], [0,1], [1,0], [1,1]]` (porte logique OR)
- But : Vérifier que notre implémentation fonctionne correctement

**Deuxième partie :** On optimise les poids avec l'algorithme de Widrow-Hoff
- Datasets réels : `p2_d1.txt` et `p2_d2.txt`
- But : Faire apprendre au perceptron à classifier automatiquement

<h3> Ce qu'on fait concrètement </h3>

1. **Perceptron simple** : Implémentation de base avec fonction d'activation
2. **Apprentissage** : Algorithme pour ajuster les poids automatiquement
3. **Visualisation** : Graphiques pour voir la frontière de décision




```python
import numpy as np 
import matplotlib.pyplot as plt 

def perceptron_simple(x,w,active) :
  x_aug=np.insert(x,0,1)  #inserer 1 au debut de la vecteur x par exemple  [1,x1,x2] c'est le biais

  #on va calculer la somme pondere : on effectue le produit scalaire entre ke vecteur des poids w et l'entree augmentee x_aug
  z=np.sum(w*x_aug)  #produit scalaire

  #fonction d'activation 
  if active==0 :   #fonction de seuil 
    y=-1  if z<0 else 1

  elif active==1 :   #fonction tanh : renvouie une valeur continue entre -1 et 1 selon z 
        y=np.tanh(z)   # or [exp(z) - exp(-z) ] / [ exp(z) + exp(-z) ]

  else :
    raise ValueError("active doit etre 0(sign) ou 1(tanh)")  # erreur pour eviter 
                                                             #un comportement inattendu
  return y

def afficher_resultats(x,y,w) :
   #on trace les points de chaque classe
   #on rouge les points de la classe 0 
   plt.scatter(x[y==0][:,0],x[y==0][:,1],color='red',label='Classe 0')
    #on vert ces qui ont on classe 1
   plt.scatter(x[y==1][:,0],x[y==1][:,1],color='green',label='Classe 1')
   #tracer la droite separatrice z=a*t +b
   a=-w[1]/w[2]  #pente
   b=-w[0]/w[2]  #ordonne a l'origine
   t=np.array([-0.5,1.5])  #valeurs de x1
   z=a*t+b
   
   #on va tracer la droite de separation en bleu . on representant la frentiere entre les classes selon la perceptron
   plt.plot(t,z,color='blue',label='Droite de separation')
   plt.show()





#fonction pour tester le perceptron dans le cas de OU logique
def   test_perceptron_ou_logique() :
    #deinition des donnees de ou logique
    x=np.array([[0,0],[0,1],[1,0],[1,1]])  #entrees
    y=np.array([0,1,1,1]) #Sorties attendues
    w=np.array([-0.5,1,1])
    print("teste de perceptron de ou logique")
    for i in range(len(x)):
        y_pred = perceptron_simple(x[i], w, active=0)
        print(f"Entrée: {x[i]}, Sortie attendue: {y[i]}, Sortie prédite: {y_pred}")
    
    for i in range(len(x)) :
        y_pred=perceptron_simple(x[i],w,active=0)
        print(f"Entree : {x[i]}, sorite attendue : {y[i]}  , sortie predite : {y_pred}\n")

    #fonction  pour afficher les poids et la droite separatrice
    afficher_resultats(x,y,w)


test_perceptron_ou_logique()
```

<h3> Analyse des résultats </h3>

Le perceptron simple fonctionne correctement sur la porte logique OR avec une précision de 100%. Le modèle sépare bien les données avec une droite de décision optimale qui distingue clairement les deux classes.

```python
import numpy as np
import matplotlib.pyplot as plt
def affiche_class(w,x,yd,title) :
    plt.figure()
    #tracer les points de l'ensemble d'apprentissage
    for i in range(x.shape[1]) :
        if yd[i]==1 :
            plt.scatter(x[0,i],x[1,i],color='green',
                        marker="o")  #classe 1 en vert
        else :
            plt.scatter(x[0,i],x[1,i],color='red',
                        marker="x")  #classe -1 en rouge
    
    if w is not None and w[2]!=0 :  #pour eviter la division par 0
       #tracer la droite separatrice z=a*t+b
       t=np.array([x[0,:].min(), x[0,:].max()])  #intervale pour x1
       a=-w[1]/w[2]  #pente
       b=-w[0]/w[2]  #ordonne a l'origine
       #on calcule les deux points(x1,x2) de la droite
       z=[a*t[0]+b,a*t[1]+b]
       plt.plot(t,z,color='blue',label='Droite de separation')
    plt.title(title)
    plt.axis([x[0,:].min()-1,x[0,:].max()+1,x[1,:].min()-1,x[1,:].max()+1]) #definir les limites des axes du graphique a tracer
    plt.grid(True)
    plt.show()

def apprentissage_widrow(x,yd,epochs,batch_size):
    
    #recuperer le nombre d'exemple d'apprentissage (colonne de x)
    n_samples=x.shape[1]
    w=np.random.random((x.shape[0]+1))
    erreur=[]  # une liste pour enregistrer l'erreur a chaque epoch
    #ajouter une ligne de 1 a x pour le biais donc le matrice devient de forme (3,n)
    x_biais = np.vstack((np.ones((1, n_samples)), x))
    print(x_biais)
    for epoch in range(epochs) :
        erreur_epoch=0
        #traiter les donnees par petites groupes(batches)
        for i in range(0,n_samples,batch_size) :
           # # Sélection du batch : sur les colonnes !
            x_batch = x_biais[:, i:i+batch_size]
            yd_batch = yd[i:i + x_batch.shape[1]]     # (batch_size,)
            print(x_batch.shape, yd_batch.shape)

    # Calcul des sorties prédites
            # calcul des sorties prédites
            y_pred = np.tanh(np.dot(w.T, x_batch))

            y_pred[y_pred == 0] = 1                 # éviter sortie 0

            #calcul de l'erreur "erreur batch" pourchaque element du batch
            erreur_batch=yd_batch-y_pred
            #mise a jour des poids w
            lr=1/batch_size  #learning rate
            #formule w=w+lr*delta_w
            #avec delta_w=np.dot(erreur_batch,x_batch.T)
            w = w + lr * np.dot(x_batch, erreur_batch.T)
            #on ajoute l'erreur quadratique du batch a l'erreur totale de l'epoch
            erreur_epoch+=np.sum(erreur_batch**2)
        
        #fonction pour tracer le droit separatrice et les points d'apprentissage d'une epoch
        affiche_class(w,x,yd,f"Epoch{epoch} (erreur={erreur_epoch})")

        #enregestrer  l'erreur cumulative de l'epoch dans la liste d'erreur
        erreur.append(erreur_epoch)
        if erreur_epoch==0 :
            break  #arreter l'apprentissage si l'erreur est nulle
    return w,erreur


#Test 1

with open("p2_d1.txt", "r") as f:
    lignes = [ligne.strip() for ligne in f if ligne.strip() != ""]  # ignore les lignes vides

# conversion en liste de listes de float
data = np.array([[float(val) for val in ligne.split()] for ligne in lignes])

x=data[:,:]  #les deux premieres colonnes sont les entrees
x=x  #transposer x pour avoir la forme (2,n)
print(f"xxx{x.shape}")
#les 25 premiers exemple appartient la classe -1 et les 25 derniers a la classe 1
yd=np.ones(50)
yd[:25]=-1
print("Apprentissage avec Widrow-Hoff")
w,erreur=apprentissage_widrow(x,yd,epochs=20,batch_size=5)
print("poids appris : ",w)
#erreur finale
print("erreur finale : ",erreur[-1])
```

<h3> Analyse des résultats  </h3>

L'algorithme converge efficacement après 20 époques avec une erreur finale de **0.148**, démontrant une séparation linéaire satisfaisante des données. Les poids optimisés **[0.144, 3.660, -0.439]** permettent une classification correcte des deux classes avec une frontière de décision stable.

# Perceptron multicoches

<h2>Perceptron multicouches - Test de validation </h2>

Dans cette section, nous implémentons une architecture de réseau de neurones avec une couche cachée de 2 neurones, sans algorithme de rétropropagation.

**Test de validation :**
- Entrée de test : **x = [1, 1]**
- Sortie attendue : **0.905**
- Architecture : 2 entrées → 2 neurones cachés → 1 sortie
- Fonction d'activation : Sigmoïde

```python
import numpy as np

def multiperceptron(x,w1,w2) :
    #Ajouter le biais a l'entree     #____w a une matrice de 3 lignes et 2 colonnes , chaque colonne correspond a une neurones cachee
                                     #______________
    x_biais=np.insert(x,0,1)
    print(x_biais)
    #calcule de la sortie de la couche cachee
    z_h1=np.dot(w1[:,0].T,x_biais)  # sortie de la premiere neurone de couche cachee
    z_h2=np.dot(w1[:,1].T,x_biais)   # sortie de la deuxieme neurones de couche cachee

    #application de la fonction d'activation(sigmoide)
    a_h1=1/(1+np.exp(-z_h1))
    a_h2=1/(1+np.exp(-z_h2))

    #creer un vecteur d'entree pour la couche de sortie
    a_h=np.array([1,a_h1,a_h2]) # ajouter le biais poue la couche de sortie
    #calcul de la sortie du neurones de la couche de sortie
    z_out=np.dot(w2.T,a_h)
    y=1/(1+np.exp(-z_out))

    return y


#On va tester le multiperceptron sur un exemple simple 
x=np.array([1 ,1])
w_1=np.array([[-0.5,0.5],
               [2,1],
               [-1,0.5]])         

w_2=np.array([2,-1,1])

y_out=multiperceptron(x,w_1,w_2)
print("sortie du multiperceptron : ",y_out)



```

```python
# 1.3.2 Apprentissage du multiperceptron

def multiperceptron_widrow(x, yd, epochs, batch_size):
    # On initialise les poids de façon fixe pour avoir des résultats reproductibles
    w_1 = np.array([[-0.5, 0.5],
                    [2, 1],
                    [-1, 0.5]])         
    w_2 = np.array([2, -1, 1])

    alpha = 0.5    # taux d'apprentissage
    erreur = []    # pour stocker les erreurs à chaque étape
    n_samples = x.shape[1]

    # boucle principale d'apprentissage
    for epoch in range(epochs):
        # on traite les données par petits groupes
        for i in range(0, n_samples, batch_size):
            x_batch = x[:, i:i+batch_size]
            yd_batch = yd[i:i+batch_size]
            
            total_erreur = 0

            # pour chaque exemple dans le batch
            for j in range(x_batch.shape[1]):
                # on ajoute le biais à l'entrée
                x_biais = np.insert(x_batch[:, j], 0, 1)
               
                # calcul des sorties de la couche cachée
                z_h1 = np.dot(w_1[:, 0].T, x_biais)
                z_h2 = np.dot(w_1[:, 1].T, x_biais)

                # fonction d'activation sigmoïde
                a_h1 = 1/(1 + np.exp(-z_h1))
                a_h2 = 1/(1 + np.exp(-z_h2))

                # entrée pour la couche de sortie (avec biais)
                a_h = np.array([1, a_h1, a_h2])

                # calcul de la sortie finale
                z_out = np.dot(w_2.T, a_h)
                y_pred = 1/(1 + np.exp(-z_out))

                # calcul de l'erreur
                erreur_sample = yd_batch[j] - y_pred
                total_erreur += erreur_sample**2

                # mise à jour des poids (version simplifiée)
                w_2 = w_2 + alpha * erreur_sample * a_h
                
                # mise à jour des poids de la couche cachée
                delta_h = erreur_sample * w_2[1:] * a_h[1:] * (1 - a_h[1:])
                w_1[:, 0] = w_1[:, 0] + alpha * delta_h[0] * x_biais
                w_1[:, 1] = w_1[:, 1] + alpha * delta_h[1] * x_biais

            erreur.append(total_erreur/batch_size)

        # affichage du progrès
        print(f"Époque {epoch+1}/{epochs}, Erreur: {total_erreur/batch_size:.4f}")

        # on s'arrête si l'erreur devient très petite
        if total_erreur < 1e-6:
            break
            
    return w_1, w_2, erreur
```
